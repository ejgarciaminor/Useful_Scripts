```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
months = np.arange(1, 13)
traffic_2023 = np.random.randint(20000, 40000, size=12)
clicks_2023 = np.random.randint(25000, 35000, size=12)
clicks_2024 = np.random.randint(20000, 40000, size=12)

# Crear el gráfico
fig, ax = plt.subplots()

# Área sombreada (2023 - total de tráfico)
ax.fill_between(months, traffic_2023, color='#DDEBF7', alpha=0.5, label='2023 - total de tráfico')

# Barras (2023 - clicks al tráfico)
ax.bar(months, clicks_2023, color='#1F4E78', label='2023 - clicks al tráfico')

# Línea (2024 - clicks al tráfico)
ax.plot(months, clicks_2024, color='#548235', label='2024 - clicks al tráfico')

# Etiquetas y leyenda
ax.set_xlabel('Meses')
ax.set_ylabel('Unidades vendidas')
ax.legend()

plt.show()

```
