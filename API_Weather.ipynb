```python
import requests
from pymeteosource.api import Meteosource
from pymeteosource.types import tiers, sections, langs, units
import pandas as pd
import datetime
today = datetime.date.today()
today = today.strftime("%Y-%m-%d")

# pip install pymeteosource
# pip install pymeteosource[pandas]

API_KEY = '3zwarn7ip5qrtgt3j8f1ne9natizvnsaqm05llya'
TIER = tiers.FREE

meteosource = Meteosource(API_KEY, TIER)
cities = ['madrid', 'barcelona', 'valencia']
daily_df = pd.DataFrame()

for city in cities:
    forecast = meteosource.get_point_forecast(
        place_id=city,
        sections=[sections.DAILY],
        lang=langs.ENGLISH,
        units=units.METRIC,
        tz='Europe/Madrid'
    )

    # Convertir los datos diarios a un DataFrame de pandas
    data = forecast.daily.to_pandas()

    # Añadir una columna con el nombre de la ciudad
    data['city'] = city

    # Añadir una columna con la fecha correspondiente
    data['date'] = data.index
    data['date_prediction'] = today

    # Concatenar al DataFrame principal
    daily_df = pd.concat([daily_df, data], ignore_index=True)

# Mostrar las primeras filas del DataFrame
daily_df.head(10)
```
